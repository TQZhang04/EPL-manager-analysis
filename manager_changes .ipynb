```python
data=[]
import time
for i in range(1992,2023):
    year=str(i)
    print(year,"start")

    params = {
        'saison_id': year,
    }



    response = requests.get(
        'https://www.transfermarkt.us/premier-league/trainerwechsel/wettbewerb/GB1/plus/1',
        params=params,
        cookies=cookies,
        headers=headers,
    )

    html_string=response.text


    soup = BeautifulSoup(html_string, 'html.parser')
    tables = soup.find_all('table', {'class': "items"})[0]
    
    tbody = tables.find_all('tbody')[0]
    tr=tbody.find_all('tr',recursive=False)[1:]
    for t in tr:
        init={}
        td=t.find_all('td',recursive=False)
        if len(td)!=12:
            continue
            
        club = td[0].find_all('img')[0].get('alt', '')
        init["club"]=club
        
        Coach = td[1].find_all('img')[0].get('alt', '')
        init["Coach"]=Coach
    
        init["Matchday"]=td[2].get_text(strip=True)
    
        init["Rank"]=td[3].get_text(strip=True)
    
        init["PPG"]=td[4].get_text(strip=True)
    
        try:
            last_match_teams_rank=td[5].find_all('span',recursive=False)[0].get_text(strip=True)
        except:
            last_match_teams_rank=""
            
        last_match_teams=f" {last_match_teams_rank} ".join([x.get('title', '') for x in td[5].find_all('a',recursive=False)])
        init["last_match"]=last_match_teams
            
        init["last_match_teams_rank"]=last_match_teams_rank
        
        init["Leaving date"]=td[6].get_text(strip=True)
    
        init["Days in charge"]=td[7].get_text(strip=True)
    
        init["Successor"] = td[8].find_all('img')[0].get('alt', '')
    
        init["Rank1"] = td[9].get_text(strip=True)
        init["PPG1"]=td[10].get_text(strip=True)
        
        try:
            first_match_teams_rank=td[11].find_all('span',recursive=False)[0].get_text(strip=True)
        except:
            first_match_teams_rank=""
            
        first_match_teams=f" {first_match_teams_rank} ".join([x.get('title', '') for x in td[5].find_all('a',recursive=False)])
        init["first_match"]=first_match_teams

        init["year"]=year
        
        data.append(init)
    print(year,"end")
        
    
    
```

```python
df = pd.DataFrame(data)
df.to_csv('output.csv', index=False)
```
